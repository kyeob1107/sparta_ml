```python
#지하철 이용자 부분 파악범위 조절 가능하게 설정 추가
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import folium
import folium
from folium.plugins import MarkerCluster

#print("pandas version: ", pd.__version__)
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
pd.set_option('display.max_row', 500)
pd.set_option('display.max_columns', 100)
file_path1 = 'F:\스파르타강의\프로젝트 데이터\정리한것\시각화데이터/2023 역별 평균승객수 순위.csv' #파일위치
file_path2 = 'F:\스파르타강의\프로젝트 데이터\정리한것\시각화데이터/청년비중20위점수.csv' #파일위치
file_path3 = 'F:\스파르타강의\프로젝트 데이터\정리한것\시각화데이터/여성청년비중20위점수.csv' #파일위치
file_path4 = 'F:\스파르타강의\프로젝트 데이터\정리한것\시각화데이터/station_coordinate.csv' #파일위치
encoding = 'euc_kr'
passenger_df = pd.read_csv(file_path1, encoding=encoding)
youth_df = pd.read_csv(file_path2, encoding=encoding)
woman_df = pd.read_csv(file_path3, encoding=encoding)
passenger_df['노선_역명']=passenger_df[['노선명','역명']].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
youth_df['노선_역명']=youth_df[['노선명','역명']].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
woman_df['노선_역명']=woman_df[['노선명','역명']].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)

#수원제외(서울내 지하철이 아니라서)
suwon_rank = passenger_df[passenger_df.노선_역명 == '경부선_수원']['총승객수 순위'].values[0] #수원등수기억
passenger_df=passenger_df[passenger_df.노선_역명!='경부선_수원'] #수원제외
passenger_df.loc[passenger_df['총승객수 순위'] > suwon_rank, '총승객수 순위'] -= 1 #순위 리매칭


#지하철 이용자 확인 순위 범위 지정부터 배점 방식까지 설정
max_score=40 #최고점수
limit_rank=20 #현재 몇등까지 점수를 부여할 것인가
way = 1 #어느 방식으로 점수부여?
#등수별 등차수열 점수부여 방식1
if way == 1:
    passenger_df['점수'] = max_score - (passenger_df['총승객수 순위'] - 1) * (max_score/limit_rank)
    passenger_df['점수'] = passenger_df['점수'].clip(lower=0)  # 음수인 경우 0으로 클립
    #0점인 부분 제외해주는 작업
    passenger_df = passenger_df[passenger_df.점수 > 0]

#등수별로 그룹 나눠서 그룹별 점수 부여 방식2
elif way == 2:
    # 등수별로 그룹을 나누어 점수 부여
    #좀더 세세한 점수 설정을 원하면 추가 수정필요
    passenger_df['그룹'] = pd.cut(passenger_df['총승객수 순위'], bins=[0, 10, 20, 30, 40, np.inf],
                                labels=['A', 'B', 'C', 'D', 'E'], right=False)
    # 각 그룹에 대한 점수 부여
    passenger_df['점수'] = passenger_df['그룹'].map({'A': 40, 'B': 30, 'C': 20, 'D': 10, 'E': 0})
    #0점인 부분 제외해주는 작업
    passenger_df = passenger_df[passenger_df.그룹 != 'E']
#끝


#이용자 수 그래프
#지하철 이용자수 20위까지 자르기
pas_20=passenger_df[passenger_df['총승객수 순위']<21] #20위까지 커트
pas_20=pas_20.reset_index(drop=True)
# 필요한 역들을 리스트에 정리
line2_stations = ['홍대입구', '구로디지털단지', '신림', '신도림', '을지로입구', '서울대입구', '성수', '신촌', '건대입구']
gangnam3_stations = ['고속터미널']
both_stations = ['잠실', '강남', '삼성', '역삼', '선릉', '사당']

# '구분' 열 추가
pas_20['구분'] = '그 외'
pas_20.loc[pas_20['역명'].isin(line2_stations), '구분'] = '2호선'
pas_20.loc[pas_20['역명'].isin(gangnam3_stations), '구분'] = '강남3구'
pas_20.loc[pas_20['역명'].isin(both_stations), '구분'] = '2호선 & 강남3구'

pas_20=pas_20.sort_values(by=['점수'], ascending=[True]) #그래프 값큰게 위로 오게
# 'pas_20'의 순위 순서대로 그리기
plt.figure(figsize=(20, 12))
plt.rc('font', size=20)

# 각 구분에 대한 hatch 패턴 및 color를 딕셔너리로 정의
styles = {'2호선': {'hatch': '/', 'color': '#FF66A6', 'edgecolor': '#FFB3D2'},
          '강남3구': {'hatch': '|', 'color': '#FFB3D2', 'edgecolor': '#FFD9E8'},
          '2호선 & 강남3구': {'hatch': 'x', 'color': '#FE006B', 'edgecolor': '#FF66A6'},
          '그 외': {'hatch': '', 'color': 'gray', 'edgecolor': 'gray'}}

for i, row in pas_20.iterrows(): #pas_20에 대해 인덱스와 그 행에 대한 값 가져오기
    category = row['구분']
    style = styles.get(category, {})  # 해당하는 구분의 스타일 정보를 가져옴
    hatch_pattern = style.get('hatch', '')  # 해당하는 구분의 hatch 패턴을 가져옴
    color = style.get('color', 'gray')  # 해당하는 구분의 color를 가져옴
    edgecolor = style.get('edgecolor','gray') # 해당하는 구분의 edgecolor를 가져옴
    bar = plt.barh(row['노선_역명'], row['점수'], color=color, edgecolor=edgecolor, label=f'{category}',
                   hatch=hatch_pattern)

# handles, labels를 사용하여 범례 정보를 얻어옴
handles, labels = plt.gca().get_legend_handles_labels()

# 중복된 항목을 방지하기 위해 이미 추가된 범례 항목을 제거하고 새롭게 범례를 추가
unique_labels = list(set(labels))
unique_handles = [handles[labels.index(label)] for label in unique_labels]

plt.legend(handles=unique_handles, labels=unique_labels, loc='best', fontsize=15)  # handles 매개변수에 막대 그래프 객체 리스트 전달
grid_lines = plt.grid(axis='x', linestyle='--', alpha=0.7) #그리드에 대한 설정
plt.gca().set_axisbelow(True)  # 그리드 그래프 뒤로
plt.tight_layout()  # 플롯 레이아웃 자동으로 조절
image_path = 'F:\스파르타강의\프로젝트 데이터\정리한것\시각화데이터/' #저장경로 지정
#plt.savefig(image_path+'지하철점수.png', transparent = True) #png로 저장
plt.show()

#청년비중 그래프
# 필요한 역들을 리스트에 정리
line2_stations = ['화양동', '신촌동', '서교동', '사당제1동']
gwanak_stations = []
both_stations = ['신림동', '낙성대동', '청룡동', '서원동', '행운동', '대학동', '신사동', '서림동', '조원동', '인헌동']

# '구분' 열 추가
youth_df['구분'] = '그 외'
youth_df.loc[youth_df['동'].isin(line2_stations), '구분'] = '2호선'
youth_df.loc[youth_df['동'].isin(gwanak_stations), '구분'] = '관악구'
youth_df.loc[youth_df['동'].isin(both_stations), '구분'] = '2호선 & 관악구'
youth_df=youth_df.sort_values(by=['점수'], ascending=[True]) #그래프 값큰게 위로 오게
plt.figure(figsize=(20, 20))
plt.rc('font', size=20)

# 각 구분에 대한 hatch 패턴 및 color를 딕셔너리로 정의
styles = {'2호선': {'hatch': '/', 'color': '#FF66A6', 'edgecolor': '#FFB3D2'},
          '관악구': {'hatch': '|', 'color': '#FFB3D2', 'edgecolor': '#FFD9E8'},
          '2호선 & 관악구': {'hatch': 'x', 'color': '#FE006B', 'edgecolor': '#FF66A6'},
          '그 외': {'hatch': '', 'color': 'gray', 'edgecolor': 'gray'}}

for i, row in youth_df.iterrows():
    category = row['구분']
    style = styles.get(category, {})  # 해당하는 구분의 스타일 정보를 가져옴
    hatch_pattern = style.get('hatch', '')  # 해당하는 구분의 hatch 패턴을 가져옴
    color = style.get('color', 'gray')  # 해당하는 구분의 color를 가져옴
    edgecolor = style.get('edgecolor','gray') # 해당하는 구분의 edgecolor를 가져옴
    bar = plt.barh(row['노선_역명'], row['점수'], color=color, edgecolor=edgecolor, label=f'{category}',
                   hatch=hatch_pattern)

# handles, labels를 사용하여 범례 정보를 얻어옴
handles, labels = plt.gca().get_legend_handles_labels()

# 중복된 항목을 방지하기 위해 이미 추가된 범례 항목을 제거하고 새롭게 범례를 추가
unique_labels = list(set(labels))
unique_handles = [handles[labels.index(label)] for label in unique_labels]

plt.legend(handles=unique_handles, labels=unique_labels, loc='best', fontsize=15)  # handles 매개변수에 막대 그래프 객체 리스트 전달
grid_lines = plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().set_axisbelow(True)  # 그리드 그래프 뒤로
plt.tight_layout()  # 플롯 레이아웃 자동으로 조절
image_path = 'F:\스파르타강의\프로젝트 데이터\정리한것\시각화데이터/'
plt.savefig(image_path+'청년비중점수.png', transparent = True)
plt.show()

# #여성청년비중 그래프
# 필요한 역들을 리스트에 정리
line2_stations = ['화양동', '신촌동', '서교동', '사당제1동']
gwanak_stations = []
both_stations = ['신림동', '낙성대동', '청룡동', '서원동', '행운동', '대학동', '신사동', '서림동', '조원동', '인헌동']

# '구분' 열 추가
woman_df['구분'] = '그 외'
woman_df.loc[woman_df['동'].isin(line2_stations), '구분'] = '2호선'
woman_df.loc[woman_df['동'].isin(gwanak_stations), '구분'] = '관악구'
woman_df.loc[woman_df['동'].isin(both_stations), '구분'] = '2호선 & 관악구'

woman_df=woman_df.sort_values(by=['점수'], ascending=[True]) #그래프 값큰게 위로 오게

plt.figure(figsize=(20, 20))
plt.rc('font', size=20)

# 각 구분에 대한 hatch 패턴 및 color를 딕셔너리로 정의
styles = {'2호선': {'hatch': '/', 'color': '#FF66A6', 'edgecolor': '#FFB3D2'},
          '관악구': {'hatch': '|', 'color': '#FFB3D2', 'edgecolor': '#FFD9E8'},
          '2호선 & 관악구': {'hatch': 'x', 'color': '#FE006B', 'edgecolor': '#FF66A6'},
          '그 외': {'hatch': '', 'color': 'gray', 'edgecolor': 'gray'}}

for i, row in woman_df.iterrows():
    category = row['구분']
    style = styles.get(category, {})  # 해당하는 구분의 스타일 정보를 가져옴
    hatch_pattern = style.get('hatch', '')  # 해당하는 구분의 hatch 패턴을 가져옴
    color = style.get('color', 'gray')  # 해당하는 구분의 color를 가져옴
    edgecolor = style.get('edgecolor','gray') # 해당하는 구분의 edgecolor를 가져옴
    bar = plt.barh(row['노선_역명'], row['점수'], color=color, edgecolor=edgecolor, label=f'{category}',
                   hatch=hatch_pattern) #막대에 대한 설정들

# handles, labels를 사용하여 범례 정보를 얻어옴
handles, labels = plt.gca().get_legend_handles_labels()

# 중복된 항목을 방지하기 위해 이미 추가된 범례 항목을 제거하고 새롭게 범례를 추가
unique_labels = list(set(labels))
unique_handles = [handles[labels.index(label)] for label in unique_labels]

plt.legend(handles=unique_handles, labels=unique_labels, loc='best', fontsize=15)  # handles 매개변수에 막대 그래프 객체 리스트 전달
grid_lines = plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().set_axisbelow(True)  # 그리드 그래프 뒤로
plt.tight_layout()  # 플롯 레이아웃 자동으로 조절
image_path = 'F:\스파르타강의\프로젝트 데이터\정리한것\시각화데이터/'
#plt.savefig(image_path+'여성청년비중점수.png', transparent = True)
plt.show()

you_group=youth_df.sort_values(by=['노선_역명'], ascending=[True])
you_group=you_group.groupby(['노선_역명'])['점수'].sum().reset_index()
won_group=woman_df.sort_values(by=['노선_역명'], ascending=[True])
won_group=won_group.groupby(['노선_역명'])['점수'].sum().reset_index()
pas_df=passenger_df[['노선_역명','점수']]
pas_df.rename(columns={'점수':'지하철점수'},inplace=True)
you_df=you_group[['노선_역명','점수']]
you_df.rename(columns={'점수':'청년점수'},inplace=True)
won_df=won_group[['노선_역명','점수']]
won_df.rename(columns={'점수':'여성청년점수'},inplace=True)
result_name=pd.concat([pas_df,you_df,won_df],ignore_index=True)
result_name=result_name.sort_values(by=['노선_역명'], ascending=[True]) #오름차순정렬
line_names=pd.DataFrame(result_name['노선_역명'].unique(),columns=['노선_역명'])

result1=pd.merge(line_names,pas_df,how='left', on='노선_역명')
result2=pd.merge(result1,you_df,how='left',on='노선_역명')#ignore_index=True)
result3=pd.merge(result2,won_df,how='left',on='노선_역명')#ignore_index=True)
#위에서 점수 부여중 숫자아니고 numpy array형태의 숫자로 들어가서 숫자로 바꿔주기
result3['지하철점수'] = result3['지하철점수'].astype(float) 
result3=result3.fillna(0) #null값 0으로
result3['총_점수']=result3['지하철점수']+result3['청년점수']+result3['여성청년점수']

#경의중앙선_신촌은 맞지 않은 역인 듯하여 제외
result3=result3[result3.노선_역명!='경의중앙선_신촌']

result3=result3.sort_values(by=['총_점수'], ascending=[False]) #내림차순정렬
result3_20=result3.nlargest(20,'총_점수',keep='all')
result3_20=result3_20.sort_values(by=['총_점수'], ascending=[True]) #그래프보기편하기 위해 오름차순
result3_20=result3_20.reset_index(drop=True)

plt.figure(figsize=(20,12))
plt.rc('font', size=20)
plt.barh(result3_20['노선_역명'], result3_20['지하철점수'], label='지하철점수', color='#FFB3D2') #lightpink #ff93c1
plt.barh(result3_20['노선_역명'], result3_20['청년점수'], left=result3_20['지하철점수'],label='청년점수', color='#FF66A6') #'hotpink' #ff4c98
plt.barh(result3_20['노선_역명'], result3_20['여성청년점수'], left=result3_20['지하철점수']+result3_20['청년점수'],label='여성청년점수', color='#FE006B') #'deeppink'
plt.legend(loc='best', fontsize=20)
#plt.grid(axis='x', linestyle=':')
grid_lines = plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().set_axisbelow(True) #그리드 그래프 뒤로
plt.tight_layout() #플롯 레이아웃 자동으로 조절해주는 메소드
image_path = 'F:\스파르타강의\프로젝트 데이터\정리한것\시각화데이터/'
#plt.savefig(image_path+'총점수(지하철점수40최대20위까지)(수원제외버전).png', transparent = True)
plt.show()

#풀리엄? 모듈 사용 간지용
c_df = pd.read_csv(file_path4, encoding=encoding)
c_df['노선_역명']=c_df[['line','name']].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
#print(c_df)
result3_20_with_coords = pd.merge(result3_20, c_df[['노선_역명', 'lat','lng']], how='left', on='노선_역명')
result3_20['lat'] = result3_20_with_coords['lat']
result3_20['lng'] = result3_20_with_coords['lng']

m = folium.Map(location=[37.5536067,126.9674308],zoom_start=10)
for n in result3_20.index:
    radius = result3_20.loc[n,'총_점수']*10
    folium.CircleMarker([result3_20['lat'][n],result3_20['lng'][n]],
                        radius = radius/50, color='#FE006B', fill=True).add_to(m)
m #얘는 따로 출력해주는 게 없는 듯해서 jupyter notebook방식 활용하여 마지막줄 실행되는 것 활용
```
